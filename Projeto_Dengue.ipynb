# Análise da Epidemia de Dengue no Brasil (2014-2024)

## 1.Introdução

A dengue faz parte de um grupo de doenças denominadas arboviroses, que se caracterizam por serem causadas por vírus transmitidos por vetores artrópodes. No Brasil, o vetor da dengue é a fêmea do mosquito Aedes aegypti (significa "odioso do Egito"). Os vírus dengue (DENV) estão classificados cientificamente na família Flaviviridae e no gênero Flavivirus. Até o momento são conhecidos quatro sorotipos – DENV-1, DENV-2, DENV-3 e DENV-4 –, que apresentam distintos materiais genéticos (genótipos) e linhagens.

As evidências apontam que o mosquito tenha vindo nos navios que partiam da África com escravos. No Brasil, a primeira epidemia documentada clínica e laboratorialmente ocorreu em 1981-1982, em Boa Vista (RR). Após quatro anos, em 1986, ocorreram epidemias atingindo o estado do Rio de Janeiro e algumas capitais da região Nordeste. Desde então, a dengue vem ocorrendo de forma continuada (endêmica), intercalando-se com a ocorrência de epidemias, geralmente associadas à introdução de novos sorotipos em áreas indenes (sem transmissão) e/ou alteração do sorotipo predominante, acompanhando a expansão do mosquito vetor.

Aspectos como a urbanização, o crescimento desordenado da população, o saneamento básico deficitário e os fatores climáticos mantêm as condições favoráveis para a presença do vetor, com reflexos na dinâmica de transmissão desses arbovírus. A dengue possui padrão sazonal, com aumento do número de casos e o risco para epidemias, principalmente entre os meses de outubro de um ano a maio do ano seguinte.

## 2.Sinais e Sintomas

A infecção por dengue pode ser assintomática ou apresentar um espectro de manifestações clínicas que varia desde formas leves até quadros graves. Geralmente, o primeiro sintoma é a febre alta (39°C a 40°C) de início abrupto, com duração de dois a sete dias, associada a dor de cabeça, dores no corpo e articulações, prostração, fraqueza, dor atrás dos olhos e erupções cutâneas.

Formas graves da doença podem incluir dor abdominal intensa, vômitos persistentes e sangramentos de mucosas, sinais que exigem atenção médica imediata.

## 3.Definição do Problema

Apesar de ser uma doença endêmica com padrão sazonal conhecido, a análise histórica dos dados de saúde pública revela uma tendência de crescimento preocupante que constitui o problema central deste estudo.

Entre 2014 e 2024, o Brasil testemunhou um aumento de aproximadamente 342% no número de casos prováveis de dengue, saltando de cerca de 1,45 milhão para mais de 6,42 milhão de registros anuais, segundo dados do Ministério da Saúde.

Este cenário de crescimento sustentado, para além das flutuações epidêmicas, torna fundamental a investigação dos fatores que contribuem para a contínua expansão da doença no país.

### 6.1.Processo de Análise (KDD)

O projeto foi estruturado seguindo as etapas do processo de ***Knowledge Discovery in Databases*** **(KDD)**, garantindo uma abordagem metódica desde a seleção dos dados até a interpretação final. As etapas foram aplicadas da seguinte forma:

1.  **Seleção:** Escolha dos conjuntos de dados de arboviroses do DATASUS, complementados por dados populacionais do Censo 2022 (IBGE) e dados climatológicos (ECMWF/ERA5), com foco no período de 2014 a 2024.
2.  **Pré-processamento:** Limpeza inicial das tabelas brutas, tratamento de valores ausentes, padronização de nomes de colunas e verificação da consistência dos registros.
3.  **Transformação:** Asta foi a etapa mais crucial e complexa do projeto. As principais transformações realizadas foram:
    * **Estimativa Mensal por Proporção Sazonal:** Aplicação da técnica para desagregar os dados anuais de casos por município, criando a granularidade `Município-Ano-Mês`, que foi a base para a maioria das análises.
    * **Criação de Métricas de Risco:** Cálculo de novas colunas (features) como a `Taxa de Incidência` por 100.000 habitantes, a `Taxa de Letalidade` e a `Taxa de Risco de Hospitalização`.
    * **Consolidação de Fontes:** Junção (merge) dos dados epidemiológicos com os dados populacionais e climáticos para criar os DataFrames finais de análise.
4.  **Mineração de Dados:** Aplicação de técnicas de estatística descritiva e visualização de dados para identificar padrões e tendências nas análises temporal, sazonal, geográfica, demográfica e de gravidade.
5.  **Interpretação e Avaliação:** Análise crítica dos resultados para validar as hipóteses levantadas e responder à pergunta norteadora.

## 8.Pré-Processamento e Feature Engineering

Com os dados brutos explorados, esta etapa é crucial para limpar, transformar e enriquecer as informações, preparando-as para a análise. O principal desafio é que nossos dados vieram de fontes agregadas e em diferentes níveis de granularidade (anual, municipal, etc.). Nosso objetivo é consolidar tudo em uma **única tabela granular**, no formato `Município-Ano-Mês`.

Para alcançar essa granularidade, que não existe nos dados brutos, aplicaremos uma técnica de engenharia de features que chamaremos de **Proportional Disaggregation**. O processo consiste em duas etapas principais:

1.  **Cálculo do Perfil Sazonal:** Primeiramente, calculamos uma tabela de proporções que representa o comportamento sazonal histórico da dengue no Brasil (ou seja, qual porcentagem de casos tende a ocorrer em janeiro, fevereiro, março, etc.).

2.  **Alocação Proporcional:** Em seguida, utilizamos esse perfil sazonal para **distribuir (alocar) o total de casos anuais de cada município** ao longo dos 12 meses.

Este passo é fundamental não apenas para viabilizar as análises temporais e geográficas, mas também para nos permitir, na sequência, calcular e integrar as proporções demográficas (idade, gênero, escolaridade), criando a base de dados rica e completa que será o motor da nossa Análise Exploratória (EDA).

```python
# --- 1. Tratamento da Tabela de Escolaridade ---
print("Processando: Infeccoes_Escolaridade...")
df_escolaridade_tidy = df_escolaridade_raw.copy()
id_vars_esc = ['Ano']
value_vars_esc = ['Em branco/IGN', 'Analfabeto', 'Ensino Fundamental', 'Ensino Médio', 'Ensino Superior']
df_escolaridade_tidy = pd.melt(df_escolaridade_tidy, id_vars=id_vars_esc, value_vars=value_vars_esc, var_name='escolaridade', value_name='qtd_casos')
df_escolaridade_tidy['qtd_casos'] = pd.to_numeric(df_escolaridade_tidy['qtd_casos'], errors='coerce')
df_escolaridade_tidy = df_escolaridade_tidy.dropna(subset=['qtd_casos'])
df_escolaridade_tidy['qtd_casos'] = df_escolaridade_tidy['qtd_casos'].astype(int)
print("-> Tabela de Escolaridade normalizada e limpa.")
display(df_escolaridade_tidy.head())
```

```python
# --- 2. Tratamento da Tabela de Faixa Etária ---
print("Processando: Infeccoes_Faixa_Etaria...")
df_faixa_etaria_tidy = df_faixa_etaria_raw.copy()
id_vars_fx = ['Ano']
value_vars_fx = ['Em branco/IGN', '<1 Ano', '1-4', '5-9', '10-14', '15-19', '20-39', '40-59', '60-64', '65-69', '70-79', '80 e +']
df_faixa_etaria_tidy = pd.melt(df_faixa_etaria_tidy, id_vars=id_vars_fx, value_vars=value_vars_fx, var_name='faixa_etaria', value_name='qtd_casos')
df_faixa_etaria_tidy['qtd_casos'] = pd.to_numeric(df_faixa_etaria_tidy['qtd_casos'], errors='coerce')
df_faixa_etaria_tidy = df_faixa_etaria_tidy.dropna(subset=['qtd_casos'])
df_faixa_etaria_tidy['qtd_casos'] = df_faixa_etaria_tidy['qtd_casos'].astype(int)
print("-> Tabela de Faixa Etária normalizada e limpa.")
display(df_faixa_etaria_tidy.head())
```

```python
# --- 3. Tratamento da Tabela de Gênero ---
print("Processando: Infeccoes_Genero...")
df_genero_tidy = df_genero_raw.copy()
id_vars_gen = ['Ano']
value_vars_gen = ['Em branco/IGN', 'Masculino', 'Feminino']
df_genero_tidy = pd.melt(df_genero_tidy, id_vars=id_vars_gen, value_vars=value_vars_gen, var_name='genero', value_name='qtd_casos')
df_genero_tidy['qtd_casos'] = pd.to_numeric(df_genero_tidy['qtd_casos'], errors='coerce')
df_genero_tidy = df_genero_tidy.dropna(subset=['qtd_casos'])
df_genero_tidy['qtd_casos'] = df_genero_tidy['qtd_casos'].astype(int)
print("-> Tabela de Gênero normalizada e limpa.")
display(df_genero_tidy.head())
```

```python
# --- 4. Tratamento da Tabela de Hospitalização ---
print("Processando: Infeccoes_Hospitalizacao...")
df_hospitalizacao_tidy = df_hospitalizacao_raw.copy()
df_hospitalizacao_tidy = df_hospitalizacao_tidy.rename(columns={'Ano': 'ano', 'Hospitalizacao': 'hospitalizacao_status', 'Qtd_Infeccoes': 'qtd_casos'})
df_hospitalizacao_tidy['qtd_casos'] = pd.to_numeric(df_hospitalizacao_tidy['qtd_casos'], errors='coerce')
df_hospitalizacao_tidy = df_hospitalizacao_tidy.dropna(subset=['qtd_casos'])
df_hospitalizacao_tidy['qtd_casos'] = df_hospitalizacao_tidy['qtd_casos'].astype(int)
print("-> Tabela de Hospitalização limpa e padronizada.")
display(df_hospitalizacao_tidy.head())
```

```python
# --- ETAPA A: Preparar a tabela de proporções mensais ---
print("Iniciando a preparação da tabela de proporções mensais...")

# Cria uma cópia do DataFrame original para garantir que os dados brutos não sejam alterados.
df_proporcoes_mensais = df_ano_mes_raw.copy()

# Padroniza o nome da coluna de ano para minúsculo.
df_proporcoes_mensais = df_proporcoes_mensais.rename(columns={'Ano': 'ano'})

# Garante que a coluna 'ano' seja numérica, removendo linhas com valores inválidos (ex: 'TOTAL').
df_proporcoes_mensais['ano'] = pd.to_numeric(df_proporcoes_mensais['ano'], errors='coerce')
df_proporcoes_mensais = df_proporcoes_mensais.dropna(subset=['ano'])
df_proporcoes_mensais['ano'] = df_proporcoes_mensais['ano'].astype(int)

# Converte todas as colunas de meses para um formato numérico, tratando erros.
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
for mes in meses:
    df_proporcoes_mensais[mes] = pd.to_numeric(df_proporcoes_mensais[mes], errors='coerce').fillna(0)

# Calcula o total de casos para cada ano somando os valores de todos os meses.
df_proporcoes_mensais['total_anual'] = df_proporcoes_mensais[meses].sum(axis=1)

# Calcula a proporção (percentual) de cada mês em relação ao total do ano.
for mes in meses:
    prop_col = f'prop_{mes.lower()}'
    # A divisão direta de colunas (vetorizada) calcula a proporção para todas as linhas de uma vez.
    df_proporcoes_mensais[prop_col] = df_proporcoes_mensais[mes] / df_proporcoes_mensais['total_anual']

# Limpa qualquer resultado inválido (resultado de divisão por zero) preenchendo com 0.
df_proporcoes_mensais = df_proporcoes_mensais.fillna(0)

# Cria uma lista com os nomes das colunas de proporção que queremos manter.
colunas_sazonais = ['ano']
for m in meses:
    colunas_sazonais.append(f'prop_{m.lower()}')

# Cria o DataFrame final desta etapa, contendo apenas o ano e as proporções mensais.
df_proporcoes_anuais = df_proporcoes_mensais[colunas_sazonais]

print("-> Tabela de proporções de sazonalidade criada com sucesso.")
display(df_proporcoes_anuais.head())
```

```python
# --- ETAPA B: Preparar a tabela de municípios ---
print("Preparando a tabela de municípios (limpeza e renomeação)...")

df_municipio_tidy = df_ano_municipio_raw.copy()

# Renomeia as colunas para um padrão mais limpo e fácil de usar.
df_municipio_tidy = df_municipio_tidy.rename(columns={
    'Município de notificação': 'municipio',
    'Ano': 'ano',
    'Qtd_Infeccoes': 'total_casos_anual'
})

# Remove o código numérico do início do nome de cada município.
df_municipio_tidy['municipio'] = df_municipio_tidy['municipio'].astype(str).str.split(' ', n=1).str[1]

# Garante que as colunas numéricas tenham o tipo de dado correto (inteiro).
df_municipio_tidy['ano'] = pd.to_numeric(df_municipio_tidy['ano'], errors='coerce').dropna().astype(int)
df_municipio_tidy['total_casos_anual'] = pd.to_numeric(df_municipio_tidy['total_casos_anual'], errors='coerce').fillna(0).astype(int)

print("-> Tabela de municípios limpa com sucesso.")
display(df_municipio_tidy.head())
```

```python
# --- ETAPA C: Junção e Alocação Proporcional ---
print("Juntando totais dos municípios com as proporções e alocando casos...")

# Junta as duas tabelas usando a coluna 'ano' como chave de conexão.
df_merged_base = pd.merge(df_municipio_tidy, df_proporcoes_anuais, on='ano')

# Para cada um dos 12 meses...
for mes in meses:
    prop_col = f'prop_{mes.lower()}'
    # ...cria uma nova coluna com a estimativa de casos daquele mês.
    df_merged_base[mes] = (df_merged_base['total_casos_anual'] * df_merged_base[prop_col]).round().astype(int)

print("-> Casos anuais alocados em estimativas mensais com sucesso.")
display(df_merged_base.head())
```

```python
# --- ETAPA D: Normalização e Ordenação Cronológica ---
print("Normalizando a tabela final para o formato longo...")

# "Derrete" a tabela, transformando as 12 colunas de meses em uma única coluna 'mes'.
df_final_granular = pd.melt(df_merged_base, id_vars=['municipio', 'ano'], value_vars=meses, var_name='mes', value_name='casos_estimados')

# Remove as linhas onde a estimativa de casos resultou em zero para otimizar o dataset.
df_final_granular = df_final_granular[df_final_granular['casos_estimados'] > 0]

# --- Ordenação Cronológica dos Meses ---
# Define a ordem correta dos meses para garantir a ordenação cronológica.
ordem_dos_meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']

# Converte a coluna 'mes' para o tipo Categórico, informando ao pandas a ordem correta.
df_final_granular['mes'] = pd.Categorical(df_final_granular['mes'], categories=ordem_dos_meses, ordered=True)

# Ordena o DataFrame final de forma correta: por município, ano e, agora, pelo mês cronológico.
df_final_granular = df_final_granular.sort_values(by=['municipio', 'ano', 'mes'])

print("\nBase granular 'df_final_granular' criada e ordenada com sucesso!")
display(df_final_granular.head())
```

```python
# --- 1. Preparando Proporções de Gênero ---
print("Processando tabela de Gênero...")

# Faz uma cópia para trabalhar
df = df_genero_tidy.copy()
index_col = 'Ano'
columns_col = 'genero'
values_col = 'qtd_casos'
prefix = 'gen'

# Calcula o total de casos para cada ano
df['total_anual'] = df.groupby(index_col)[values_col].transform('sum')

# Calcula a proporção de cada categoria dentro do seu ano
df['proporcao'] = df[values_col] / df['total_anual']

# Pivota a tabela para ter as proporções como colunas
df_pivotado = df.pivot_table(index=index_col, columns=columns_col, values='proporcao').fillna(0)

# Converte o índice ('Ano') de volta para uma coluna e padroniza o nome
df_pivotado = df_pivotado.reset_index()
df_pivotado = df_pivotado.rename(columns={index_col: 'ano'})

# Adiciona o prefixo 'gen_' às colunas de categoria para evitar conflitos
cols_to_rename = df_pivotado.columns.drop('ano')
new_names = {col: f"{prefix}_{col}" for col in cols_to_rename}
df_prop_genero = df_pivotado.rename(columns=new_names)

print("-> Tabela de proporções de Gênero criada com sucesso.")
display(df_prop_genero.head())
```

```python
# --- 2. Preparando Proporções de Escolaridade ---
print("\nProcessando tabela de Escolaridade...")

# A lógica é EXATAMENTE a mesma, apenas mudamos os DataFrames e nomes de entrada
df = df_escolaridade_tidy.copy()
index_col = 'Ano'
columns_col = 'escolaridade'
values_col = 'qtd_casos'
prefix = 'esc'

df['total_anual'] = df.groupby(index_col)[values_col].transform('sum')
df['proporcao'] = df[values_col] / df['total_anual']
df_pivotado = df.pivot_table(index=index_col, columns=columns_col, values='proporcao').fillna(0)
df_pivotado = df_pivotado.reset_index()
df_pivotado = df_pivotado.rename(columns={index_col: 'ano'})
cols_to_rename = df_pivotado.columns.drop('ano')
new_names = {col: f"{prefix}_{col}" for col in cols_to_rename}
df_prop_escolaridade = df_pivotado.rename(columns=new_names)

print("-> Tabela de proporções de Escolaridade criada com sucesso.")
display(df_prop_escolaridade.head())
```

```python
# --- 3. Preparando Proporções de Faixa Etária ---
print("\nProcessando tabela de Faixa Etária...")

df = df_faixa_etaria_tidy.copy()
index_col = 'Ano'
columns_col = 'faixa_etaria'
values_col = 'qtd_casos'
prefix = 'fx'

df['total_anual'] = df.groupby(index_col)[values_col].transform('sum')
df['proporcao'] = df[values_col] / df['total_anual']
df_pivotado = df.pivot_table(index=index_col, columns=columns_col, values='proporcao').fillna(0)
df_pivotado = df_pivotado.reset_index()
df_pivotado = df_pivotado.rename(columns={index_col: 'ano'})
cols_to_rename = df_pivotado.columns.drop('ano')
new_names = {col: f"{prefix}_{col}" for col in cols_to_rename}
df_prop_faixa_etaria = df_pivotado.rename(columns=new_names)

print("-> Tabela de proporções de Faixa Etária criada com sucesso.")
display(df_prop_faixa_etaria.head())
```

```python
# --- 4. Preparando Proporções de Hospitalização ---
print("\nProcessando tabela de Hospitalização...")

df = df_hospitalizacao_tidy.copy()
# A coluna de ano já foi renomeada para 'ano', então usamos 'ano' aqui
index_col = 'ano'
columns_col = 'hospitalizacao_status'
values_col = 'qtd_casos'
prefix = 'hosp'

df['total_anual'] = df.groupby(index_col)[values_col].transform('sum')
df['proporcao'] = df[values_col] / df['total_anual']
df_pivotado = df.pivot_table(index=index_col, columns=columns_col, values='proporcao').fillna(0)
df_pivotado = df_pivotado.reset_index()
df_pivotado = df_pivotado.rename(columns={index_col: 'ano'})
cols_to_rename = df_pivotado.columns.drop('ano')
new_names = {col: f"{prefix}_{col}" for col in cols_to_rename}
df_prop_hospitalizacao = df_pivotado.rename(columns=new_names)

print("-> Tabela de proporções de Hospitalização criada com sucesso.")
display(df_prop_hospitalizacao.head())
```

```python
print("Iniciando a junção final para criar a base de dados de análise...")

# Inicia-se o dataframe de análise a partir de uma cópia da base granular.
df_analise = df_final_granular.copy()

# Realiza a junção sequencial com cada uma das tabelas de proporções demográficas.
# O método 'left' join garante a manutenção de todos os registros da base principal.
df_analise = pd.merge(df_analise, df_prop_genero, on='ano', how='left')
df_analise = pd.merge(df_analise, df_prop_escolaridade, on='ano', how='left')
df_analise = pd.merge(df_analise, df_prop_faixa_etaria, on='ano', how='left')
df_analise = pd.merge(df_analise, df_prop_hospitalizacao, on='ano', how='left')

# Identifica todas as colunas que são de tipo numérico.
colunas_numericas = df_analise.select_dtypes(include=np.number).columns

# Preenche possíveis valores nulos com 0 apenas nas colunas numéricas,
# preservando colunas de outros tipos (como a categórica 'mes').
df_analise[colunas_numericas] = df_analise[colunas_numericas].fillna(0)

print("\nDataFrame de análise final ('df_analise') criado com sucesso!")
print("Esta é a base de dados completa que será usada para todos os gráficos e análises.")
display(df_analise.head())
```

```python
print("--- Verificando a Integridade dos Dados Após a Junção Final ---")

# 1. Calcula o total de casos por ano ANTES da junção (a partir do df_final_granular)
total_antes_do_merge = df_final_granular.groupby('ano')['casos_estimados'].sum()
total_antes_do_merge.name = 'Total_Antes_do_Merge'

# 2. Calcula o total de casos por ano DEPOIS da junção (a partir do df_analise)
total_depois_do_merge = df_analise.groupby('ano')['casos_estimados'].sum()
total_depois_do_merge.name = 'Total_Depois_do_Merge'

# 3. Juntar os dois resultados em uma única tabela para uma comparação lado a lado
df_validacao_final = pd.concat([total_antes_do_merge, total_depois_do_merge], axis=1)

# 4. Calcular a diferença entre os totais (deve ser zero)
df_validacao_final['Diferenca'] = df_validacao_final['Total_Antes_do_Merge'] - df_validacao_final['Total_Depois_do_Merge']

print("\nComparação dos totais de 'casos_estimados' por ano:")
display(df_validacao_final)
```

### 8.1.Validação Final do Processo de Engenharia de Dados

Após a complexa etapa de Desagregação Proporcional, é fundamental realizar uma verificação de consistência para garantir a integridade dos nossos dados. O objetivo é confirmar que o processo de estimativa não criou nem destruiu casos de forma significativa.

A tabela a seguir compara o total de casos anuais da fonte de dados original (`Total_Original`) com a soma de nossas estimativas mensais para cada ano (`Total_Estimado_Final`).

É esperado que exista uma pequena diferença, resultante dos múltiplos arredondamentos feitos em cada cálculo de estimativa mensal. Para quantificar essa discrepância e avaliar a precisão do nosso método, adicionamos uma coluna de **Diferença Percentual**. Valores muito próximos de zero indicam um alto grau de precisão e validam a abordagem utilizada.

```python
print("--- Validação da Consistência dos Totais Anuais ---")

# Calcula o total de casos por ano a partir do DataFrame final de análise.
total_final_estimado = df_analise.groupby('ano')['casos_estimados'].sum()

# Calcula o total de casos por ano a partir da fonte de dados original.
total_original = df_proporcoes_mensais.set_index('ano')['total_anual']

# Cria um DataFrame para a comparação dos resultados.
df_validacao_completa = pd.DataFrame({
    'Total_Original': total_original,
    'Total_Estimado_Final': total_final_estimado
})

# Calcula a diferença absoluta (magnitude do erro) para ser sempre positiva.
df_validacao_completa['Diferenca_Absoluta'] = (df_validacao_completa['Total_Estimado_Final'] - df_validacao_completa['Total_Original']).abs()

# Calcula a diferença percentual, baseada na diferença absoluta.
df_validacao_completa['Diferenca_Percentual'] = (
    (df_validacao_completa['Diferenca_Absoluta'] / df_validacao_completa['Total_Original'])
).fillna(0) * 100

# --- FORMATAÇÃO DA COLUNA DE PERCENTUAL ---
df_validacao_completa['Diferenca_Percentual'] = df_validacao_completa['Diferenca_Percentual'].apply(
    lambda x: f'{x:.2f}%'
)

print("\nComparação entre os dados Originais e o resultado Estimado Final:")

# Exibe a tabela final de validação.
display(df_validacao_completa)
```

Portanto, a precipitação funciona como um dos principais **gatilhos ambientais** que precedem e impulsionam a temporada de dengue no Brasil.

Logo, investigaremos a Taxa de Análise de Incidência por Escolaridade.

```python
# --- Análise da Taxa de Incidência por Escolaridade (com dados do IBGE) ---

# 1. Criando o DataFrame do IBGE diretamente com os dados do Censo 2022
# Estes dados foram extraídos da Tabela 9605 do SIDRA/IBGE.
print("Criando DataFrame com dados do IBGE (Censo 2022)...")
dados_censo_2022 = {
    'Escolaridade_IBGE': [
        'Sem instrução e fundamental incompleto',
        'Fundamental completo e médio incompleto',
        'Médio completo e superior incompleto',
        'Superior completo',
        'Não determinado'
    ],
    'Populacao_Total': [
        57445741,
        25953177,
        59756087,
        21010777,
        286284
    ]
}

df_ibge = pd.DataFrame(dados_censo_2022)
print("Dados do IBGE carregados com sucesso!")
display(df_ibge)


# 2. Agrupar os dados do IBGE para corresponder às nossas categorias
# O IBGE agrupa 'Analfabeto' com 'Fundamental Incompleto'. Vamos usar essa categoria.
# A categoria "Sem instrução e fundamental incompleto" do IBGE engloba nosso "Analfabeto" e parte do "Fundamental".
pop_analfabeto_e_fundamental = df_ibge[
    df_ibge['Escolaridade_IBGE'].str.contains("Sem instrução|fundamental incompleto", case=False)
]['Populacao_Total'].sum()

# Vamos atribuir o restante do fundamental para a categoria de fundamental completo
pop_fundamental_completo = df_ibge[
    df_ibge['Escolaridade_IBGE'].str.contains("Fundamental completo", case=False)
]['Populacao_Total'].sum()

pop_medio = df_ibge[
    df_ibge['Escolaridade_IBGE'].str.contains("Médio completo", case=False)
]['Populacao_Total'].sum()

pop_superior = df_ibge[
    df_ibge['Escolaridade_IBGE'].str.contains("Superior completo", case=False)
]['Populacao_Total'].sum()

# 3. Criar um DataFrame com a população total por categoria, alinhado ao nosso 'dados_esc'
# Para simplificar, vamos agrupar 'Analfabeto' e 'Ensino Fundamental' na população
# já que os casos nesses grupos são sociologicamente próximos e o IBGE os agrupa.
populacao_agrupada = pd.DataFrame({
    'Escolaridade': ['Analfabeto', 'Ensino Fundamental', 'Ensino Médio', 'Ensino Superior'],
    'Populacao_Total': [
        pop_analfabeto_e_fundamental, # Usando o grupo maior para as duas categorias
        pop_analfabeto_e_fundamental + pop_fundamental_completo,
        pop_medio,
        pop_superior
    ]
})


# 4. Juntar os dados de CASOS de dengue com os dados de POPULAÇÃO do IBGE
# 'dados_esc' é o DataFrame que criamos na análise anterior com o total de casos
df_taxa_incidencia = pd.merge(dados_esc, populacao_agrupada, on='Escolaridade')

# 5. Calcular a Taxa de Incidência por 100.000 habitantes
df_taxa_incidencia['Taxa_Incidencia_100k'] = (df_taxa_incidencia['Total de Casos Estimados'] / df_taxa_incidencia['Populacao_Total']) * 100000

print("\nDataFrame final com a Taxa de Incidência:")
display(df_taxa_incidencia)


# --- Gráfico de Barras: Taxa de Incidência por Escolaridade ---
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")

barplot_taxa = sns.barplot(data=df_taxa_incidencia, x='Taxa_Incidencia_100k', y='Escolaridade', palette='plasma_r', order=ordem_grafico_esc)

plt.title('Taxa de Incidência de Dengue por 100.000 Habitantes por Escolaridade (2014-2024)', fontsize=16, pad=20)
plt.xlabel('Casos por 100.000 Habitantes', fontsize=14, labelpad=15)
plt.ylabel('Nível de Escolaridade', fontsize=14, labelpad=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.tight_layout()
plt.show()
```

O gráfico de taxa de incidência recalibra nossa compreensão e revela as seguintes conclusões críticas:

1.  **O Paradoxo da Baixa Incidência no Grupo Analfabeto:** De forma contraintuitiva, o grupo "Analfabeto" apresenta a menor taxa de incidência. A explicação mais plausível para este fenômeno não é uma menor exposição ao vírus, mas sim uma **severa subnotificação**. Indivíduos neste grupo podem ter menor acesso ao sistema de saúde formal, o que leva a menos diagnósticos e notificações oficiais. Este achado pode indicar uma importante limitação nos dados de saúde pública.

2.  **O Epicentro do Risco - Ensino Médio:** O grupo com Ensino Médio se destaca negativamente. Além de possuir o maior volume de casos, também apresenta a **maior taxa de incidência**, posicionando-se como o grupo de maior risco e impacto para a dengue no Brasil.

3.  **Risco Elevado no Ensino Superior:** A taxa de incidência no grupo com Ensino Superior é a segunda maior, superando a do Ensino Fundamental, o que não ocorreu na análise apenas por números absolutos. Isso reforça que, apesar dos fatores de proteção socioeconômicos (que reduzem o volume total de casos), o risco para este grupo permanece elevado, provavelmente devido à sua forte presença na faixa etária de maior risco (20-39 anos) e à concentração em grandes centros urbanos.

##### Conclusão da Análise Demográfica

A análise demográfica, ao cruzar diferentes perspectivas, conclui que o perfil mais impactado pela dengue é a **população adulta em idade produtiva (20-59 anos)**, com destaque para o grupo com **Ensino Médio**, que apresenta o maior risco relativo.

Como identificamos previamente que **o público com maior número de infecções eram os adultos**, mas que os **grupos escolares com mais casos** da doença eram o **Ensino Médio e Fundamental**, levou ao questionamento:
**Por que o Ensino Superior não é o grupo com mais casos, já que é representado pela população adulta, essa mesma que apresenta mais números de casos?**
Esse questionamento levantou **duas hipóteses**:


1.   A análise deveria ser feita por Taxa de Incidência, para nos dar números mais reais.
2.   Existem muitos adultos que ainda estão no Nívem Médio, já que o país apresenta um baixo nível de escolaridade.

A primeira hipótese foi respondida pela nova ótica da Taxa de Incidência, em vez do cálculo por valores absolutos. Já a segunda, encontramos mais dificuldades por questões de disponibilidade dos dados, e por não ser o cerne do projeto.


A investigação também levanta uma hipótese crítica sobre a **subnotificação de casos em populações de baixíssima instrução**, sugerindo que a vulnerabilidade deste grupo pode ser ainda maior do que os dados oficiais aparentam.

### 9.6.Análise de Gravidade

Após traçarmos o perfil demográfico dos afetados pela dengue, a próxima questão fundamental é:

**Qual a gravidade do impacto da doença? O número de óbitos e a taxa de letalidade se mantiveram estáveis ou apresentaram tendências preocupantes?**

Nesta seção, vamos quantificar a severidade da epidemia analisando a proporção de casos que evoluíram para hospitalização. Este indicador é crucial para entender a pressão sobre o sistema de saúde e o sofrimento humano gerado pela doença.

#### 9.6.1.O Impacto da Hospitalização

Analisaremos o total de casos estimados que geraram ou não hospitalização entre 2014 e 2024.

```python
# --- Análise de Gravidade: Casos de Hospitalização ---

# 1. Calcular o total de casos estimados para cada categoria de hospitalização
casos_hospitalizados_sim = np.sum(df_analise['casos_estimados'] * df_analise['hosp_Sim'])
casos_hospitalizados_nao = np.sum(df_analise['casos_estimados'] * df_analise['hosp_Não'])

# 2. Criar um DataFrame para a visualização
dados_hosp = pd.DataFrame({
    'Condição': ['Hospitalizados', 'Não Hospitalizados'],
    'Total de Casos Estimados': [casos_hospitalizados_sim, casos_hospitalizados_nao]
})

# 3. Calcular as proporções para anotar o gráfico
total_geral_casos = dados_hosp['Total de Casos Estimados'].sum()
dados_hosp['Percentual'] = (dados_hosp['Total de Casos Estimados'] / total_geral_casos) * 100

print("Dados de Hospitalização Compilados:")
display(dados_hosp)


# --- Gráfico de Barras: Comparativo de Hospitalização ---
plt.figure(figsize=(10, 7))
sns.set_style("whitegrid")

# Usando uma paleta de cores que pode sugerir alerta/neutro
barplot_hosp = sns.barplot(data=dados_hosp, x='Condição', y='Total de Casos Estimados', palette=['#c94a49', '#7e7e7e'])

# Adicionando anotações precisas (número absoluto e percentual) em cada barra
for index, row in dados_hosp.iterrows():
    barplot_hosp.annotate(f"{int(row['Total de Casos Estimados']):,} | {row['Percentual']:.2f}%".replace(',', '.'),
                           (index, row['Total de Casos Estimados']),
                           ha='center', va='bottom',
                           xytext=(0, 5),
                           textcoords='offset points',
                           fontsize=14,
                           color='black')

# Títulos, rótulos e formatação
plt.title('Total de Casos Estimados: Hospitalizados vs. Não Hospitalizados (2014-2024)', fontsize=18, pad=20)
plt.xlabel('Condição do Paciente', fontsize=14, labelpad=15)
plt.ylabel('Total de Casos Estimados (em milhões)', fontsize=14, labelpad=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.ylim(0, dados_hosp['Total de Casos Estimados'].max() * 1.15) # Espaço extra para as anotações
plt.ticklabel_format(style='plain', axis='y')

plt.tight_layout()
plt.show()
```

##### Análise do Gráfico de Impacto da Hospitalização

A análise inicial de gravidade quantifica o impacto direto da dengue no sistema de saúde. O gráfico acima revela que, embora a grande maioria dos casos (95.25%) não resulte em internação, a proporção de **4.75% de casos hospitalizados** representa um volume absoluto massivo. Ao longo de uma década, isso se traduz em mais de 500 mil internações, demonstrando a enorme pressão que a dengue exerce sobre os recursos hospitalares do país.

#### 9.6.2.Aprofundando a Análise de Gravidade: Foco em Hospitalizações por Faixa Etária
Para validar as informações sobre a suscetibilidade de certos grupos, como os adolescentes, aprofundamos a análise focando nos dados de hospitalização. Investigamos o volume de internações por faixa etária, tanto no panorama geral (2014-2024) quanto no ano específico do surto (2024), e calculamos a taxa de risco para entender a vulnerabilidade real de cada grupo.

Os adolescentes são tidos como o grupo que representa maior número de hospitalizações, como já mencionado anteriormente, em virtude de não haver tido contato com outros genótipos da doença, o que impossibilita a imunização total.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# --- 1. CÁLCULO DO NUMERADOR: Total de Hospitalizações por Faixa Etária (2014-2024) ---
print("Calculando o total de hospitalizações por faixa etária...")

# Mapeamento dos nomes das colunas para os rótulos do gráfico
mapa_fx_etaria = {
    'fx_<1 Ano': '<1 Ano', 'fx_1-4': '1-4 anos', 'fx_5-9': '5-9 anos',
    'fx_10-14': '10-14 anos', 'fx_15-19': '15-19 anos', 'fx_20-39': '20-39 anos',
    'fx_40-59': '40-59 anos', 'fx_60-64': '60-64 anos', 'fx_65-69': '65-69 anos',
    'fx_70-79': '70-79 anos', 'fx_80 e +': '80 e + anos'
}
colunas_fx = list(mapa_fx_etaria.keys())

# Cálculo para o período completo
lista_hosp_total = []
for coluna in colunas_fx:
    hosp_estimados = (df_analise['casos_estimados'] * df_analise[coluna] * df_analise['hosp_Sim']).sum()
    lista_hosp_total.append({
        'faixa_etaria': mapa_fx_etaria[coluna],
        'hospitalizacoes_estimadas': hosp_estimados
    })
df_hosp_total = pd.DataFrame(lista_hosp_total)


# --- 2. CÁLCULO DO DENOMINADOR: População por Faixa Etária (Fonte: IBGE, Censo 2022) ---
print("Processando dados de população do IBGE...")
dados_pop_faixa_etaria = {
    'faixa_etaria_ibge': ['Menor de 1 ano','1 a 4 anos','5 a 9 anos','10 a 14 anos','15 a 19 anos','20 a 24 anos','25 a 29 anos','30 a 34 anos','35 a 39 anos','40 a 44 anos','45 a 49 anos','50 a 54 anos','55 a 59 anos','60 a 64 anos','65 a 69 anos','70 a 74 anos','75 a 79 anos','80 anos e mais'],
    'populacao_total': [2155781, 9593170, 12693918, 13463806, 14594229, 15343825, 15998114, 16335111, 15923985, 15918231, 14249069, 12836287, 10760431, 8754605, 7150066, 5634563, 3965931, 4615858]
}
df_pop_fx_ibge = pd.DataFrame(dados_pop_faixa_etaria)

# Função para agrupar as faixas etárias do IBGE para corresponder às nossas
def agrupar_faixa_etaria(faixa_ibge):
    faixa_ibge = str(faixa_ibge).strip()
    if faixa_ibge == 'Menor de 1 ano': return '<1 Ano'
    if faixa_ibge == '1 a 4 anos': return '1-4 anos'
    if faixa_ibge == '5 a 9 anos': return '5-9 anos'
    if faixa_ibge == '10 a 14 anos': return '10-14 anos'
    if faixa_ibge == '15 a 19 anos': return '15-19 anos'
    if any(s in faixa_ibge for s in ['20 a 24', '25 a 29', '30 a 34', '35 a 39']): return '20-39 anos'
    if any(s in faixa_ibge for s in ['40 a 44', '45 a 49', '50 a 54', '55 a 59']): return '40-59 anos'
    if faixa_ibge == '60 a 64 anos': return '60-64 anos'
    if faixa_ibge == '65 a 69 anos': return '65-69 anos'
    if any(s in faixa_ibge for s in ['70 a 74', '75 a 79']): return '70-79 anos'
    if faixa_ibge == '80 anos e mais': return '80 e + anos'
    return 'Outro'

df_pop_fx_ibge['faixa_etaria'] = df_pop_fx_ibge['faixa_etaria_ibge'].apply(agrupar_faixa_etaria)
populacao_agrupada_fx = df_pop_fx_ibge.groupby('faixa_etaria')['populacao_total'].sum().reset_index()


# --- 3. CÁLCULO FINAL E PLOTAGEM ---
# Juntar dados de hospitalização e população
df_taxa_hosp = pd.merge(df_hosp_total, populacao_agrupada_fx, on='faixa_etaria')

# Calcular a Taxa de Hospitalização por 100.000 habitantes
df_taxa_hosp['taxa_hosp_100k'] = (df_taxa_hosp['hospitalizacoes_estimadas'] / df_taxa_hosp['populacao_total']) * 100000
df_taxa_hosp = df_taxa_hosp.sort_values(by='taxa_hosp_100k', ascending=False)

print("\nDataFrame final com a Taxa de Risco de Hospitalização:")
display(df_taxa_hosp[['faixa_etaria', 'taxa_hosp_100k']])

# Plotar o Gráfico Final
plt.figure(figsize=(18, 10))
sns.set_style("whitegrid")
ax = sns.barplot(data=df_taxa_hosp, x='taxa_hosp_100k', y='faixa_etaria', palette='viridis_r')
plt.title('Taxa de Risco de Hospitalização por Dengue (por 100.000 Habitantes)', fontsize=18, pad=20)
plt.xlabel('Hospitalizações por 100.000 Habitantes da Faixa Etária', fontsize=12)
plt.ylabel('Faixa Etária', fontsize=12)
plt.tight_layout()
plt.show()
```

##### Interpretação da Taxa de Risco de Hospitalização

Este gráfico revela a história que os números absolutos escondem. Ao normalizar pela população de cada grupo, confirmamos que o **risco individual** de um caso de dengue se agravar a ponto de necessitar de hospitalização é mais pronunciado em **adolescentes (10-19 anos) e bebês (<1 ano)**. Este achado valida a hipótese de que fatores imunológicos e a dinâmica de circulação de novos sorotipos tornam esses grupos mais vulneráveis.

#### 9.6.3.Análise de Óbitos e Taxa de Letalidade

Para aprofundar a análise de gravidade, investigamos a evolução do número de óbitos e da taxa de letalidade ao longo do tempo. A taxa de letalidade, em particular, é um indicador crítico da severidade da doença e da capacidade de resposta do sistema de saúde a cada ano.

```python
# --- Início da Análise de Óbitos e Letalidade ---

# Passo 1 (CORRIGIDO): Limpeza robusta da coluna de ano no df_obitos_raw
# Converte a coluna para numérico. Se encontrar um erro (como um texto vazio), transforma em NaN (Not a Number).
df_obitos_raw['Ano notificação'] = pd.to_numeric(df_obitos_raw['Ano notificação'], errors='coerce')

# Remove qualquer linha que tenha ficado com o ano nulo (NaN), que são as linhas problemáticas
df_obitos_raw.dropna(subset=['Ano notificação'], inplace=True)

# Agora, com a coluna limpa, podemos converter para inteiro com segurança
df_obitos_raw['Ano notificação'] = df_obitos_raw['Ano notificação'].astype(int)


# Garante que a coluna 'ano' em df_analise também seja do tipo inteiro
df_analise['ano'] = df_analise['ano'].astype(int)


# Renomeia e trata a tabela de óbitos
df_obitos_anual = df_obitos_raw.rename(columns={
    'Ano notificação': 'ano',
    'Óbito pelo agravo notificado': 'obitos'
})
df_obitos_anual = df_obitos_anual[['ano', 'obitos']]
print("Dados de óbitos anuais carregados e tratados com sucesso.")

# Calcula o total de casos estimados por ano a partir do df_analise
casos_por_ano = df_analise.groupby('ano')['casos_estimados'].sum().reset_index()

# Junta as tabelas de casos e óbitos
df_gravidade = pd.merge(casos_por_ano, df_obitos_anual, on='ano')

# Calcula a Taxa de Letalidade em percentual
df_gravidade['taxa_letalidade_percent'] = (df_gravidade['obitos'] / df_gravidade['casos_estimados']) * 100

print("\nDataFrame 'df_gravidade' criado com sucesso e pronto para as visualizações.")
display(df_gravidade)
```

```python
# --- Gráfico 1: Evolução do Número Absoluto de Óbitos por Ano ---
plt.figure(figsize=(15, 7))
sns.set_style("whitegrid")

ax1 = sns.barplot(data=df_gravidade, x='ano', y='obitos', palette='Reds_d', alpha=0.9)

for p in ax1.patches:
    ax1.annotate(format(p.get_height(), ',.0f').replace(',', '.'),
                   (p.get_x() + p.get_width() / 2., p.get_height()),
                   ha = 'center', va = 'center', xytext = (0, 9),
                   textcoords = 'offset points', fontsize=12)

plt.title('Evolução do Número de Óbitos por Dengue no Brasil (2014-2024)', fontsize=18, pad=20)
plt.xlabel('Ano da Notificação', fontsize=12)
plt.ylabel('Número de Óbitos', fontsize=12)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
ax1.set(ylim=(0, df_gravidade['obitos'].max() * 1.15))
plt.tight_layout()
plt.show()
```

```python
# --- Gráfico 2: Evolução da Taxa de Letalidade por Ano ---
plt.figure(figsize=(15, 7))
sns.set_style("whitegrid")

ax2 = sns.lineplot(data=df_gravidade, x='ano', y='taxa_letalidade_percent', marker='o', color='black', linewidth=2.5)

for index, row in df_gravidade.iterrows():
    plt.text(row['ano'], row['taxa_letalidade_percent'] + 0.005, f"{row['taxa_letalidade_percent']:.2f}%",
             ha='center', va='bottom', fontsize=11, color='black', fontweight='bold')

plt.title('Evolução da Taxa de Letalidade da Dengue no Brasil (2014-2024)', fontsize=18, pad=20)
plt.xlabel('Ano da Notificação', fontsize=12)
plt.ylabel('Taxa de Letalidade (%)', fontsize=12)

# Define os ticks do eixo X para serem exatamente os anos do nosso dataframe, evitando erros
plt.xticks(ticks=df_gravidade['ano'])
plt.xticks(fontsize=11, rotation=45, ha='right')
plt.yticks(fontsize=11)

from matplotlib.ticker import FuncFormatter
ax2.yaxis.set_major_formatter(FuncFormatter(lambda y, _: f'{y:.2f}%'))
ax2.set(ylim=(0, df_gravidade['taxa_letalidade_percent'].max() * 1.2))
plt.tight_layout()
plt.show()
```

##### Análise dos Óbitos e da Letalidade

Os gráficos acima revelam a faceta mais trágica da epidemia. A evolução do número absoluto de óbitos mostra picos em anos de grandes surtos, como 2015 e 2022-2023, mas nada se compara à explosão de fatalidades em **2024**, que superou 6.000 mortes, um recorde histórico.

A análise da **Taxa de Letalidade** (o percentual de casos que evoluem para óbito) é ainda mais crítica. Ela mostra que, após anos de relativa estabilidade, o risco de um caso de dengue levar à morte voltou a subir de forma preocupante a partir de 2022, atingindo o ponto mais alto da década em 2024 (0.10%). Isso pode indicar a circulação de sorotipos mais virulentos ou uma sobrecarga no sistema de saúde.
Apesar de 2017 e 2018 terem tido os menores números de casos estimados da doença, eles só ficaram atrás de 2024 em termos de taxa de letalidade.

#### 9.6.4. Análise de Letalidade entre Pacientes Hospitalizados

Para a análise final de gravidade, investigamos um indicador ainda mais específico: a taxa de letalidade dentro do universo de pacientes já hospitalizados.

Esta métrica é um indicador crítico da **severidade dos casos que demandam cuidados intensivos** e pode servir como uma avaliação indireta da capacidade de resposta do sistema hospitalar durante os surtos. A pergunta a ser respondida é: **"Dos pacientes que desenvolvem a forma grave da doença a ponto de serem internados, qual a proporção que, mesmo com o tratamento, evolui para óbito?"**. A análise da evolução desta taxa ao longo do tempo pode revelar mudanças na virulência do vírus ou na eficácia do tratamento clínico dos casos graves.

```python
# --- Nova Análise: Taxa de Letalidade entre Hospitalizados ---

# Passo 1: Calcular o total de hospitalizações por ano a partir do df_analise
# Usamos a mesma lógica da desagregação: soma(casos * proporção de hospitalização)
hospitalizados_por_ano = df_analise.groupby('ano').apply(
    lambda x: (x['casos_estimados'] * x['hosp_Sim']).sum()
).reset_index(name='total_hospitalizados')

# Passo 2: Juntar os dados de hospitalizações com os de óbitos
df_letalidade_hosp = pd.merge(hospitalizados_por_ano, df_obitos_anual, on='ano')

# Passo 3: Calcular a Taxa de Letalidade entre Hospitalizados
df_letalidade_hosp['taxa_letalidade_hosp_percent'] = (df_letalidade_hosp['obitos'] / df_letalidade_hosp['total_hospitalizados']) * 100

print("DataFrame para a análise de letalidade entre hospitalizados:")
display(df_letalidade_hosp)


# --- Gráfico 3: Evolução da Taxa de Letalidade entre Hospitalizados ---
plt.figure(figsize=(15, 7))
sns.set_style("whitegrid")

ax3 = sns.lineplot(data=df_letalidade_hosp, x='ano', y='taxa_letalidade_hosp_percent', marker='o', color='#8B0000', linewidth=2.5)

for index, row in df_letalidade_hosp.iterrows():
    plt.text(row['ano'], row['taxa_letalidade_hosp_percent'] + 0.05, f"{row['taxa_letalidade_hosp_percent']:.2f}%",
             ha='center', va='bottom', fontsize=11, color='black', fontweight='bold')

plt.title('Taxa de Letalidade da Dengue entre Pacientes Hospitalizados (2014-2024)', fontsize=18, pad=20)
plt.xlabel('Ano da Notificação', fontsize=12)
plt.ylabel('Óbitos por 100 Pacientes Hospitalizados (%)', fontsize=12)
plt.xticks(ticks=df_letalidade_hosp['ano'])
plt.xticks(fontsize=11, rotation=45, ha='right')
plt.yticks(fontsize=11)

ax3.yaxis.set_major_formatter(FuncFormatter(lambda y, _: f'{y:.2f}%'))
ax3.set(ylim=(0, df_letalidade_hosp['taxa_letalidade_hosp_percent'].max() * 1.2))
plt.tight_layout()
plt.show()
```

##### Análise da Letalidade entre Hospitalizados

Este indicador mede a gravidade dos casos que chegam ao sistema de saúde e a eficácia do tratamento hospitalar. A análise da sua evolução ao longo do tempo é fundamental para a vigilância epidemiológica.
